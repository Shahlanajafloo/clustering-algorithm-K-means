```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as dt
import sklearn.cluster as cl
```

```python
X = dt.load_iris().data
```

```python
KMN=cl.KMeans(n_clusters=3)
```

```python
KMN.fit(X)
```

```python
lables=KMN.labels_
```

```python
Center=KMN.cluster_centers_
```

```python
plt.scatter(X[:,2],X[:,3],s=15,c=lables)
plt.scatter(Center[:,2],Center[:,3],s=60,c='r',marker='x')
```
